```python
df = pd.DataFrame(datas)

df
```

```python
df1 = df[(df["Age"] > 30) & (df["Outcome"] == 1)] # 30 yaş üstü diyabet olanlar
df1
```

```python
df2 = df[(df["Age"] > 30) & (df["Outcome"] == 0)] # 30 yaş üstü diyabet olmayanlar
df2
```

```python
df3 = df[(df["Age"] < 30) & (df["Outcome"] == 1)] # 30 yaş altı diyabet olanlar
df3
```

```python
df4 = df[(df["Age"] < 30) & (df["Outcome"] == 0)] # 30 yaş altı diyabet olmayanlar
df4
```

```python
x = np.array(["Sarı","Yeşil","Siyah", "Mavi", "Kırmızı"])
y = np.array([len(df1),len(df2),len(df3),len(df4),len(df)])

plt.bar(x[0],y[0], color="yellow", label="30 yaş üstü diyabet olanlar")
plt.bar(x[1],y[1], color="green", label="30 yaş üstü diyabet olmayanlar")
plt.bar(x[2],y[2], color="black", label="30 yaş altı diyabet olanlar")
plt.bar(x[3],y[3], color="blue", label="30 yaş altı diyabet olmayanlar")
plt.bar(x[4],y[4], color="red", label="Toplam")

plt.legend()
plt.show()
```

```python
x = np.array(["Sarı","Yeşil","Kırmızı"])
y = np.array([len(df1),len(df2), len(df) - len(df3) - len(df4)])

plt.bar(x[0],y[0], color="yellow", label="30 yaş üstü diyabet olanlar")
plt.bar(x[1],y[1], color="green", label="30 yaş üstü diyabet olmayanlar")
plt.bar(x[2],y[2], color="red", label="30 Yaş üstü toplamı")

plt.legend()
plt.show()
```

```python
df1 = df[(df["Pregnancies"] > 0) & (df["Outcome"] == 1)] # En az 1 kere hamile kalmış diyabet olanlar
df1
```

```python
df2 = df[(df["Pregnancies"] > 0) & (df["Outcome"] == 0)] # En az 1 kere doğum yapmış diyabet olmayanlar
df2
```

```python
x = np.array(["Sarı","Yeşil","Kırmızı", "black"])
y = np.array([len(df1),len(df2), len(df1) + len(df2),len(df) - len(df1) - len(df2)])

plt.bar(x[0],y[0], color="yellow", label="En az 1 kere hamile kalmış diyabet olanlar")
plt.bar(x[1],y[1], color="green", label="En az 1 kere hamile kalmış diyabet olmayanlar")
plt.bar(x[2],y[2], color="red", label="En az 1 kere hamile kalanlar")
plt.bar(x[3],y[3], color="black", label="En az 1 kere bile hamile kalmayanlar")

plt.legend()
plt.show()
```

```python
df1 = df[(df["Glucose"] > df["Glucose"].mean()) & (df["Outcome"] == 1)] # Ortalamanın üstünde glikoz oranı olan ve diyabet olanlar
df1
```

```python
df2 = df[(df["Glucose"] > df["Glucose"].mean()) & (df["Outcome"] == 0)] # Ortalamanın üstünde glikoz oranı olan ve diyabet olmayanlar
df2
```

```python
df3 = df[(df["Glucose"] < df["Glucose"].mean()) & (df["Outcome"] == 1)] # Ortalamanın altında glikoz oranı olan ve diyabet olanlar
df3
```

```python
df4 = df[(df["Glucose"] < df["Glucose"].mean()) & (df["Outcome"] == 0)] # Ortalamanın altında glikoz oranı olan ve diyabet olmayanlar
df4
```

```python
y = np.array([len(df1),len(df2),len(df3),len(df4)])
mylabels = [
    "Ortalamanın üstünde glikoz oranı olan ve diyabet olanlar",
    "Ortalamanın üstünde glikoz oranı olan ve diyabet olmayanlar",
    "Ortalamanın altında glikoz oranı olan ve diyabet olanlar",
    "Ortalamanın altında glikoz oranı olan ve diyabet olmayanlar",
           ];
plt.pie(y, labels=mylabels, explode = [0.1,0.1,0.1,0.1], autopct='%1.1f%%')

plt.show()
```
